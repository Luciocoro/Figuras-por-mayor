```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('figuras_lucio.mplstyle')
```

```python
n = 2
y = []
t = np.linspace(0,10,1000)
for i in range(n):
    y.append((i+1)*np.sin(4*t))
y = np.array(y)
```

```python
fig, ax = plt.subplots()

for i in range(n):
    ax.plot(t,y[i])
```
